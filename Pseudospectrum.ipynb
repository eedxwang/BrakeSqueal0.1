```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import random
from matplotlib.colors import LogNorm
%matplotlib inline
```

```python
#generate input data for a sample input parameters
Emin=190*10**6
Emax=203*10**6
E=(Emin+Emax)/2.0
rhomin=7750
rhomax=8050
rho=(rhomin+rhomax)/2.0
L=1.0
```

```python
w1=np.pi*np.sqrt(E/rho)/2*L
```

```python
#vary all parameters randomly with mean=0 as above and variance 0.01
m=1000
sigma=0.01
pset=np.array([E,rho,L])
pset_repeat=np.reshape(np.tile(pset,m),(m,3))
X=pset_repeat
```

```python
X[:,0]=(E+1e5*np.random.randn(1,m)).flatten()
X[:,1]=(rho+1e-2*np.random.randn(1,m)).flatten()
X[:,2]=(L+1e-6*np.random.randn(1,m)).flatten()
```

```python
X
```

```python
y=[]
for i in np.arange(m):
    E,rho,L=X[i]
    w1=np.pi*np.sqrt(E/rho)/2*L
    y.append(w1)
```

```python
df = pd.DataFrame(data = X)
df.columns = ['E','rho','L']
df['frequency'] = y
df.head()
```

```python
df['frequency'].var()
```
